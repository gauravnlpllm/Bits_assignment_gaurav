```python
import os

from groq import Groq
from google.colab import userdata

client = Groq(
    api_key=userdata.get('mykey'),
)
```

```python
def set_open_params(
    model="llama3-8b-8192",
    temperature=0.7,
    max_tokens=256,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0,
):
    """Set groq parameters."""
    groq_params = {}
    groq_params['model'] = model
    groq_params['temperature'] = temperature
    groq_params['max_tokens'] = max_tokens
    groq_params['top_p'] = top_p
    groq_params['frequency_penalty'] = frequency_penalty
    groq_params['presence_penalty'] = presence_penalty
    return groq_params

# Example usage:
params = set_open_params()
print("groq parameters:", params)
```

```python
def get_completion(params, prompt):
  response = client.chat.completions.create(
      model = params['model'],
      messages = [{"role": "user", "content": prompt}],
      temperature = params['temperature'],
      max_tokens = params['max_tokens'],
      top_p = params['top_p'],
      frequency_penalty = params['frequency_penalty'],
      presence_penalty = params['presence_penalty'],
      )
  return response
```

```python
params = set_open_params()
prompt = "The sky is"
response = get_completion(params, prompt)
```

```python
response.choices[0].message.content
```
